```python
df = df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

### Now labeling Gender and Overtime because they are binary categories
### Rest we will perform OneHotEncoding

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Label encoding binary columns
binary_columns = ['Gender', 'OverTime']
for col in binary_columns:
  le = LabelEncoder()
  df[col] = le.fit_transform(df[col])
```

```python
# Defining ColumnTransformer for OneHot Encoding
categorical_columns = ['BusinessTravel',
 'Department',
 'EducationField',
 'JobRole',
 'MaritalStatus']
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns)
    ],
    remainder='passthrough'
)
```

```python
df.head()
```

```python
# Split data into X, y
X = df.drop('Attrition', axis=1)
y = df['Attrition']

# train, test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])
```

```python
model_pipeline.fit(X_train, y_train)
```

```python
model_pipeline.score(X_test, y_test)
```

```python
# If train > test -> Overfitting
# If train = test -> good
# If train and test both low -> underfitting
print('Training accuracy', model_pipeline.score(X_train, y_train))
print('Testing accuracy', model_pipeline.score(X_test, y_test))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [5, 10, None],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1,2,4]
}

grid_search = GridSearchCV(estimator=model_pipeline,
                           param_grid = param_grid,
                           cv=2,
                           n_jobs=-1,
                           verbose=1,
                           scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
np.random.seed(42)
best_model = grid_search.best_estimator_
test_score = best_model.score(X_test, y_test)
```
